# 分析
* 题目一共需要完成两个任务：
    * 判断链表是否有环
    * 环的入口在哪？


# 思路：
## 首先需要完成任务一：判断是否有环
* 判断是否有环通常用快慢指针： fast 和 slow
    * 比较简单的做法是 slow 每次走1步，而且 fast 比 slow 快一倍，也就是 fast 走两步
    * 下图是相遇的过程：

    ![jupyter](https://code-thinking.cdn.bcebos.com/gifs/141.%E7%8E%AF%E5%BD%A2%E9%93%BE%E8%A1%A8.gif)

* 为什么第一次在环中相遇，slow的 步数 是 x+y 而不是 x + 若干环的长度 + y 呢？
    * 首先slow进环的时候，fast一定是先进环来了.
    * 如果slow进环入口，fast也在环入口，那么把这个环展开成直线，就是如下图的样子：  

    ![jupyter](https://img-blog.csdnimg.cn/2021031816503266.png)
    
    * 可以看出如果slow 和 fast同时在环入口开始走，一定会在环入口3相遇，slow走了一圈，fast走了两圈。
        * 重点来了，slow进环的时候，fast一定是在环的任意一个位置，如图：

    ![jupyter](https://img-blog.csdnimg.cn/2021031816515727.png)

    * 那么fast指针走到环入口3的时候，已经走了k + n 个节点，slow相应的应该走了(k + n) / 2 个节点。

        * 因为k是小于n的（图中可以看出），所以(k + n) / 2 一定小于n。

        * 也就是说slow一定没有走到环入口3，而fast已经到环入口3了。

        * 这说明在slow开始走的那一环已经和fast相遇了

    * One more thing:   为什么fast不能跳过去呢？ 
        * fast相对于slow是一次移动一个节点，所以不可能跳过去。


## 在完成了一的基础上才能完成任务二：环的入口在哪里
* 相遇时： slow指针走过的节点数为: x + y， fast指针走过的节点数：x + y + n (y + z)，n为fast指针在环内走了n圈才遇到slow指针， （y+z）为 一圈内节点的个数A。
    * 遵循上面的 fast slow 设定，会有如下公式
        $$ 2*(x+y) = x + n*(y+z) $$
        $$ \Rightarrow x = (n-1)(x+y) + z $$

        ![jupyter](https://code-thinking-1253855093.file.myqcloud.com/pics/20220925103433.png)
        
        * 当 n=1 时，有 $ x = z $, 
            * 此时，若从头结点出发一个指针，从相遇节点也出发一个指针，这两个指针每次只走一个节点，那么当这两个指针相遇的时候就是 环形入口的节点。
        
        * 那么 n如果大于1是什么情况呢，就是fast指针在环形转n圈之后才遇到 slow指针。
            * 其实这种情况和n为1的时候 效果是一样的，一样可以通过这个方法找到 环形的入口节点，只不过，index1 指针在环里 多转了(n-1)圈，然后再遇到index2，相遇点依然是环形的入口节点。
    
    * 因此，要求出入口的位置，做法是：
        * 在相遇节点处，定义一个指针index1，在头结点处定一个指针index2。
        * 让index1和index2同时移动，每次移动一个节点， 那么他们相遇的地方就是 环形入口的节点。如下图：

        ![jupyter](https://code-thinking.cdn.bcebos.com/gifs/142.%E7%8E%AF%E5%BD%A2%E9%93%BE%E8%A1%A8II%EF%BC%88%E6%B1%82%E5%85%A5%E5%8F%A3%EF%BC%89.gif)


# 出题都是单链表

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def detectCycle(self, head: Optional[ListNode]) -> Optional[ListNode]:
        
        # 任务一：判断是否有环
        fast, slow = head, head                          # 快慢指针
        
        while fast != None and fast.next != None:        # 因为fast走的快，所以只需要验一下fast，一定不要漏了 fast.next，否则可能出现对空指针操作
            fast = fast.next.next
            slow = slow.next

            # 任务二：找到环的入口
            if fast == slow and fast != None:            # 找到环
                idx1 = fast                              # 一个指针放在相遇的位置，这里用 fast 或者 slow 都可以
                idx2 = head                              # 另一个指针放在 head 位置
                while idx1 != idx2:
                    idx1 = idx1.next                     # 注意移动速度是相同的
                    idx2 = idx2.next
                return idx1                              # return 的是 idx1 或 idx2， 别搞错了
        
        return None
```
