# 二分法

* 使用要求：
    * 有序
    * 无重复元素


* 要点：
    * 在进行二分查找的过程中，每一次边界的处理都要根据区间的定义去操作，区间的定义在整个while查找过程中都要保持不变。
    


# 思路一：
    * 区间定义：左闭右闭
    * 区间的合理性判断： left <= right 是合理区间
    * 当 num[middle] > target 时，更新下一次查找的区间为 [ left, middle - 1 ] 

```python
def dichonomy_1(nums, target):
    '''
        写法：
            1、初始化左右区间的index: left 和 right
            2、写while循环的判断条件: 根据「左闭右闭」的区间定义，循环条件为 left <= right
            3、 在while循环中，定义 mid = (left + right) // 2
            4、 根据不同情况，更新左右区间的index
    '''
    # 初始化左右区间
    left = 0
    right = len(nums) - 1
    
    # 写while循环的判断条件
    while (left <= right):
        
        # 定义 mid 变量
        mid = (left + right) //2   # 或者在C++语言里会用的防止溢出写法 mid = left + (right-left)//2
        
        # 更新区间
        if (nums[mid] > target):
            right = mid - 1        # [left , target_idx, mid, right] --> [left, mid -1]
        elif (nums[mid] < target ):
            left = mid + 1         # [left, mid, target_idx, right]  --> [mid+1, right]
        else:
            return mid
    
    return -1
    
```

```python
# test dichonomy_1 
import numpy as np

target = 10

# nums = [0,1,2,3,5,7,9,10]
nums = np.random.randint(0,100,10)
nums.sort()
print(f"nums:{nums} | target: {target}")


dichonomy_1(nums,target)
```

# 思路二：
    * 区间定义：左闭右开
    * 区间的合理性判断： left < right 是合理区间
    * 当 num[middle] > target 时，更新下一次查找的区间为 [ left, middle ]

```python
def dichonomy_2(nums, target):
    '''
        写法：
            1、初始化left 和 right
            2、写while 循环的判断条件： left < right
            3、 定义/更新 mid 变量： mid = (left + right) // 2
            4、 根据不同情况，更新左右区间的index
    '''
    # 初始化区间指针
    left = 0
    right = len(nums)           # 注意区间要包含到最后一个元素哦！ 
    
    # 写while循环的判断条件
    while (left < right):       # 左闭右开
        # 定义/更新 mid 变量 
        mid = (left + right) // 2
    
        if (nums[mid] > target):
            right = mid         # [left, target_idx, mid, right) --> [left, mid)
        elif (nums[mid] < target):
            left = mid +1       # [left, mid, target_idx, right) --> [mid+1, right)
        else: 
            return mid
    
    return -1
```

```python
# test dichonomy_2 
import numpy as np

target = 128

nums = [0,1,2,3,5,7,10,18, 128]
# nums = np.random.randint(0,100,10)
# nums.sort()

print(f"nums:{nums} | target: {target}")

dichonomy_2(nums,target)
```
