* 给你两个单链表的头节点 headA 和 headB ，请你找出并返回两个单链表相交的起始节点。如果两个链表没有交点，返回 null 。

    * 解释：相交节点的值为 2 （注意，如果两个链表相交则不能为 0）。

* 提示：

    * 0 <= skipB <= n
    * 如果 listA 和 listB 没有交点，intersectVal 为 0
    * 如果 listA 和 listB 有交点，intersectVal == listA[skipA + 1] == listB[skipB + 1]

# 分析：
* 两个链表的交点指的是： 相交位置要求节点的[指针]相等，而不单单是 val 相等
* 如果两个链表相交，那么相交节点之后的所有节点必然相同
    * 因为 每一个节点只有一个 next，相交之后不可能再分叉了

# 思路：
* 先把两个链表的长度求出来
* 再把两个链表的尾部对齐
* 如果两个链表的长度不相等，则把长的链表遍历到短链表的开始位置，然后开始对比节点的指针
    * 如果两个链表长度相等，则直接从两个链表的初始位置开始对比节点的指针
* 对比链表节点的逻辑：
    * 要对比链表的节点，就必须把 current_node 放到「待对比节点的前一个位置」，否则无法获取待指向对比节点的指针
    * 如果找到相等的指针（非空指针），则返回 current_node.next 
    * 当 current_node.next = None 时停止对比，并返回空指针


```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> ListNode:

        # 遍历两个链表，获取长度
        m = 0 
        n = 0
        current_point_A = headA
        current_point_B = headB
        while current_point_A is not None:               # m，n统计的是节点的个数，不是从0开始数的，所以current要一直走到 None，否则会少数一个
            m += 1
            current_point_A = current_point_A.next
        while current_point_B is not None:
            n += 1
            current_point_B = current_point_B.next
        
        # 把两个链表的尾部进行对齐
        temp_gap = max(m-n,n-m)
        current_point_A = headA                          # 重置遍历的起点，此时current为head，也就是 链表 index=-1 的位置
        current_point_B = headB
        if m>n:                                         # A 比较长
            while temp_gap:
                current_point_A = current_point_A.next
                temp_gap -= 1
        else:
            while temp_gap:
                current_point_B = current_point_B.next
                temp_gap -= 1
        
        # 对比两个节点
        while current_point_A != None:                   # 由于已经对齐，所以判断A和B都是相同的
            if current_point_A == current_point_B:        # 由于 head 是一个指针，所以这里判断的其实是两个指针是否相同，这个指针指向的是 current 节点的下一个位置
                return current_point_A                   # 本质上这里返回的是指向下一个节点的指针，比如 返回head指针，函数返回后得到的是head的下一个节点
            
            current_point_A = current_point_A.next
            current_point_B = current_point_B.next
        
        return None 
        
```
